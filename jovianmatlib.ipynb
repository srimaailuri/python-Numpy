```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('years')
plt.ylabel('yeild(tons per hectare)');
```

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
#marker says about circle,square.....,c--color,ls--linestyle,lw--linewidth,ms--markerstyle,mew--markeredgewidth,mec--markeredgecolor
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
#fmt='[marker][linestyle][color]'
plt.plot(years, apples,'s-b' )
plt.plot(years, oranges,'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years,oranges,'or')
plt.title("yeild of Oranges(tones per hectare)");
```

```python
#figure figsize command is used to change the shape of the graph
plt.figure(figsize=(12,6))
plt.plot(years,oranges,'or')
plt.title("yeild of Oranges(tones per hectare)");
```

## scatter plots

```python
#iris dataset is already present in seaborn module so we can load it using load_dataset
flowers_df=sns.load_dataset("iris")
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length,flowers_df.petal_width);
```

```python
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width);
```

```python
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width,hue=flowers_df.species,s=100);
```

```python
#s=size of the points
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species,
                s=100);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')

sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                hue='species',
                s=100,
               data=flowers_df);
```

## histogram

```python
# when we come to histogram we have to take only one column only 
# describe function gives you the statistics of the dataframe
flowers_df.describe()
```

```python
flowers_df.sepal_width
```

```python
plt.hist(flowers_df.sepal_width)
```

```python
#bins of unequal size
plt.hist(flowers_df.sepal_width,bins=[1,3,4,4.5]);
```

```python
import numpy as np
```

```python
setosa_df=flowers_df[flowers_df.species=='setosa']
versicolor_df=flowers_df[flowers_df.species=='versicolor']
virginica_df=flowers_df[flowers_df.species=='virginica']
```

```python
plt.hist(setosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
plt.hist(virginica_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25));
plt.legend(['setosa','Virginica']);
```

```python
# we can also stack multiple graph's
plt.hist([setosa_df.sepal_width,versicolor_df.sepal_width,virginica_df.sepal_width],bins=np.arange(2,5,0.25),stacked=True);
plt.legend(['setosa','versicolor','Virginica']);
```

```python
tips_df=sns.load_dataset('tips')
tips_df
```

```python
bill_avg=tips_df.groupby('day')[['total_bill']].mean()
```

```python
#syntax for bargraph=plt.bar(x,y)
#here in bar graph we want x axis should have day column and y axis should have totalbill column 
# which means index of the bill_avg is day column 
plt.bar(bill_avg.index,bill_avg.total_bill);
```

```python
sns.barplot(x='day', y='total_bill', data=tips_df);
```

```python
#hue argument help in comparing depending on the third column
sns.barplot(x='day', y='total_bill',hue='sex', data=tips_df);
```

```python
# we can even do horizontal bars
sns.barplot(x='total_bill',y='day',hue='sex', data=tips_df)
```

## heatmap

```python
df=sns.load_dataset("flights")
df
```

```python
plt.plot(df.passengers)
```

```python
flights_df=sns.load_dataset("flights").pivot("month","year","passengers")
flights_df
```

```python
plt.title("no. of passengers")
sns.heatmap(flights_df);
```

```python
#fmt is used for placing a minimum distance.
#annot is used to display numbers
#cmap is used for selecting the color
plt.title("no. of passengers")
sns.heatmap(flights_df,fmt="d",annot=True,cmap='Blues');
```
